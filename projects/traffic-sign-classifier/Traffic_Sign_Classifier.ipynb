## Step 2: Design and Test a Model Architecture

Designing and implementing a deep learning model that learns to recognize traffic signs. Train and test the model on the [German Traffic Sign Dataset](http://benchmark.ini.rub.de/?section=gtsrb&subsection=dataset).


### Pre-process the Data Set

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 


````python
from sklearn.utils import shuffle
import math
import cv2

def preprocessAndNormalize(rgb):
    yuv = np.empty(rgb.shape)
    # Change the RGB colorspace to YUV
    for i in range(0, rgb.shape[0]):
        yuv[i] = cv2.cvtColor(rgb[i], cv2.COLOR_RGB2YUV)
        
    # Normalize the YUV data
    yuv[:,:,:, 0] = (yuv[:,:,:, 0] - 128.0)/128.0 #Y range: 0 to 255
    yuv[:,:,:, 1] = (yuv[:,:,:, 1])/128.0 #U range: -128 to 127
    yuv[:,:,:, 2] = (yuv[:,:,:, 2])/128.0 #V range: -128 to 127
    return yuv
````

### Preprocess and Normalize training, validation, and test data

````python
# training data
X_train = preprocessAndNormalize(X_train)

# validation data
X_valid = preprocessAndNormalize(X_valid)

# training data
X_test = preprocessAndNormalize(X_test)
````

### Model Architecture

````python
import tensorflow as tf

EPOCHS = 30
BATCH_SIZE = 128

rate = 0.00075
accuracy_thres = 0.94
````

# Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

### Create the Model

````python
from tensorflow.contrib.layers import flatten

def ConvNet(x):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # SOLUTION: Layer 1: Convolutional. Input = 32x32x3. Output = 28x28x16.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 3, 16), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(16))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)
    
    # SOLUTION: Pooling. Input = 28x28x16. Output = 14x14x16.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Layer 2: Convolutional. Output = 10x10x32.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 16, 32), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(32))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)

    # SOLUTION: Pooling. Input = 10x10x32. Output = 5x5x32.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # SOLUTION: Layer 3: Convolutional. Output = 5x5x64.
    conv3_W = tf.Variable(tf.truncated_normal(shape=(1, 1, 32, 64), mean = mu, stddev = sigma))
    conv3_b = tf.Variable(tf.zeros(64))
    conv3   = tf.nn.conv2d(conv2, conv3_W, strides=[1, 1, 1, 1], padding='VALID') + conv3_b
    
    # SOLUTION: Activation.
    conv3 = tf.nn.relu(conv3)
    
    # SOLUTION: Pooling. Input = 5x5x64. Output = 5x5x64.
    conv3 = tf.nn.max_pool(conv3, ksize=[1, 2, 2, 1], strides=[1, 1, 1, 1], padding='SAME')

    # SOLUTION: Flatten. Input = 5x5x64. Output = 1600 + 800 (flattened conv3 layer concatenated with flattened conv2 layer).
    fc0   = tf.concat([flatten(conv3), flatten(conv2)], axis=1)
    
    # SOLUTION: Layer 4: Fully Connected. Input = 1600+800. Output = 512.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(2400, 512), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(512))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)
    # Dropout Regularization
    fc1 = tf.nn.dropout(fc1, keep_prob = 0.4)

    # SOLUTION: Layer 5: Fully Connected. Input = 512. Output = 84.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(512, 84), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(84))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)
    # Dropout Regularization
    fc1 = tf.nn.dropout(fc1, keep_prob = 0.4)
    
    # SOLUTION: Layer 6: Fully Connected. Input = 84. Output = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(84, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    # Create a Network parameter dict for visualization
    global network_params
    network_params = {
        "conv1": conv1,
        "conv2": conv2,
        "conv3": conv3,
        "fc0": fc0,
        "fc1": fc1,
        "fc2": fc2,
        "logits": logits
    }
    
    return logits

x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)

logits = ConvNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()
````

### Evaluate Function

````python
def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

### Tensorflow Training and Validation Session

Train and validate the model here and save the session so that it can be retrieved later to test its performance

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = n_train
    
    training_accuracy = np.float64(np.empty(EPOCHS))
    validation_accuracy = np.float64(np.empty(EPOCHS))
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        training_accuracy[i] = evaluate(X_train, y_train)    
        validation_accuracy[i] = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Training Accuracy = {:.3f}%".format(training_accuracy[i]*100))
        print("Validation Accuracy = {:.3f}%".format(validation_accuracy[i]*100))
        
        # Early termination to avoid overfitting
        if (validation_accuracy[i] >= accuracy_thres):
            EPOCHS = i+1
            break    

    saver.save(sess, './saved_sess/traffic_sign_lenet')
    print("Model saved")
````

### Plot the Training and Validation Accuracy

````python
# Load the saved model
with tf.Session() as sess:
    saver.restore(sess, './saved_sess/traffic_sign_lenet')
    plt.figure
    train_plot = plt.plot(range(EPOCHS), training_accuracy[range(EPOCHS)], 'r-', label='Training Accuracy') 
    valid_plot = plt.plot(range(EPOCHS), validation_accuracy[range(EPOCHS)], 'b-', label='Validation Accuracy')
    plt.legend()
    plt.title("Training and Validation Accuracy")
    plt.xlabel("EPOCHS")
    plt.grid(color='r', linestyle='--', linewidth=0.5);
    plt.show()
````
